```
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import numpy as np
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
```

```
iris = load_iris()
x = iris["data"][:, 3:]
log_reg = LogisticRegression()
```

```
#iris setosa  versus non iris setosa
```

```
y = (iris["target"] == 0).astype(np.int) #1 if iris setosa class else 0
```

```
def createDataSet():
  xTrain, xTest, yTrain, yTest = train_test_split(x,y, test_size = 0.3, random_state = 101)
  return xTrain,xTest,yTrain,yTest
```

```
xTrain,xTest,yTrain,yTest = createDataSet()
log_reg.fit(xTrain, yTrain)
y_predict = log_reg.predict(xTest)
```

```
y_proba = log_reg.predict_proba(xTest)
plt.plot(xTest, y_proba[:, 1], "g-"
, label="Iris-Virginica")
plt.plot(xTest, y_proba[:, 0], "b--"
, label="Not Iris-Virginica")
```

```
print(accuracy_score(yTest, y_predict, normalize= True))
```
